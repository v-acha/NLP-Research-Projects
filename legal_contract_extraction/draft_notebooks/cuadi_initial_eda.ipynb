```python
import json
from datasets import Dataset

# Define the file path (update if necessary)
file_path = "/content/drive/MyDrive/Colab Notebooks/CUAD_v1.json"

# Load JSON file
with open(file_path, "r") as f:
    cuad_data = json.load(f)

# Print structure of the first entry
print(json.dumps(cuad_data["data"][:1], indent=2))

```

#### Step 3: Convert CUAD to Hugging Face datasets Format
We need to extract:

context: The contract text.
question: The legal-related question.
answers: The extracted spans from the contract.

```python
def convert_to_hf_format(cuad_data):
    """
    Convert CUAD JSON data to Hugging Face `datasets` format.
    """
    squad_formatted_data = {
        "id": [],
        "title": [],
        "context": [],
        "question": [],
        "answers": []
    }

    for document in cuad_data["data"]:
        title = document["title"]
        for paragraph in document["paragraphs"]:
            context = paragraph["context"]
            for qa in paragraph["qas"]:
                squad_formatted_data["id"].append(qa["id"])
                squad_formatted_data["title"].append(title)
                squad_formatted_data["context"].append(context)
                squad_formatted_data["question"].append(qa["question"])

                # Handle cases where no answer exists
                if qa["is_impossible"]:
                    squad_formatted_data["answers"].append({"text": [], "answer_start": []})
                else:
                    squad_formatted_data["answers"].append({
                        "text": [ans["text"] for ans in qa["answers"]],
                        "answer_start": [ans["answer_start"] for ans in qa["answers"]]
                    })

    return Dataset.from_dict(squad_formatted_data)

# Convert CUAD to Hugging Face Dataset
cuad_hf_dataset = convert_to_hf_format(cuad_data)


```
